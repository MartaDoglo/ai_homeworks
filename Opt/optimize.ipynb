```python
from scipy import optimize
import numpy as np
from matplotlib import pyplot as plt

def bil(x):
    return (1.5 - x[0] + x[0]*x[1])**2 + (2.25 - x[0] + x[0]*x[1]**2)**2 + (2.629 - x[0] + x[0]*x[1]**3)**2
```

```python
X = np.arange(-4.5, 4.5, 0.1)
Y = np.arange(-4.5, 4.5, 0.1)
X, Y = np.meshgrid(X, Y)
Z = bil(np.array([X,Y]))

x0 = np.array([1.3, 0.7, 0.8, 1.9, 1.2])
res = optimize.minimize(bil, x0, method='BFGS')
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='coolwarm')

print("res:", res.x[0], res.x[1], res.fun)

ax.plot3D(res.x[0], res.x[1], res.fun, 'o', linewidth=5, color='red')
ax.view_init(30, 240)
plt.show()
```
